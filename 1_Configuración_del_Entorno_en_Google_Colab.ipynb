## Fundamentos de NumPy

### Creación de Arrays NumPy

NumPy, la biblioteca fundamental para computación numérica en Python, ofrece diversas formas de crear arrays multidimensionales (ndarrays).

1.  **Desde listas o tuplas de Python:** La forma más común es convertir una lista o tupla de Python en un array de NumPy usando `np.array()`.
2.  **Arrays con valores iniciales:**
    *   `np.zeros(shape)`: Crea un array lleno de ceros.
    *   `np.ones(shape)`: Crea un array lleno de unos.
    *   `np.full(shape, fill_value)`: Crea un array lleno de un valor específico.
    *   `np.empty(shape)`: Crea un array con contenido aleatorio (dependiendo del estado de la memoria).
3.  **Arrays con rangos numéricos:**
    *   `np.arange(start, stop, step)`: Crea un array con valores espaciados uniformemente dentro de un intervalo, similar a `range()` de Python.
    *   `np.linspace(start, stop, num)`: Crea un array con un número específico de valores espaciados uniformemente dentro de un intervalo.
