## Gráficos básicos con Matplotlib

### Subtask:
Crea y muestra gráficos básicos como líneas, dispersión e histogramas utilizando Matplotlib. Asegúrate de incluir etiquetas de ejes y un título.


**Reasoning**:
First, I need to import the necessary libraries, `matplotlib.pyplot` for plotting and `numpy` for generating sample data. Then, I will generate random data for line, scatter, and histogram plots. Finally, I will create each type of plot with appropriate labels and titles, and display them.



```python
import matplotlib.pyplot as plt
import numpy as np

# 2. Generate some sample data
# Data for line plot
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Data for scatter plot
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10 + np.cos(x_scatter)

# Data for histogram
data_hist = np.random.randn(1000)

# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 3. Create a basic line plot
axes[0].plot(x_line, y_line, color='blue', linestyle='-', marker='o', markersize=4)
axes[0].set_xlabel('Eje X (Línea)')
axes[0].set_ylabel('Eje Y (Línea)')
axes[0].set_title('Gráfico de Línea Simple')
axes[0].grid(True)

# 4. Create a scatter plot
axes[1].scatter(x_scatter, y_scatter, color='red', alpha=0.7)
axes[1].set_xlabel('Eje X (Dispersión)')
axes[1].set_ylabel('Eje Y (Dispersión)')
axes[1].set_title('Gráfico de Dispersión')
axes[1].grid(True)

# 5. Create a histogram
axes[2].hist(data_hist, bins=30, color='green', edgecolor='black', alpha=0.7)
axes[2].set_xlabel('Valores (Histograma)')
axes[2].set_ylabel('Frecuencia (Histograma)')
axes[2].set_title('Histograma de Datos Aleatorios')
axes[2].grid(axis='y', alpha=0.75)

# 6. Show all plots
plt.tight_layout()
plt.show()
```

## Gráficos avanzados con Matplotlib

### Subtask:
Explora funcionalidades avanzadas de Matplotlib, como subgráficos, personalización de colores y estilos, y la adición de leyendas a gráficos complejos.


**Reasoning**:
I need to generate sample data, create a figure with multiple subplots, plot a customized line graph with a legend in the first subplot, and a customized scatter plot in the second subplot, then adjust the layout and display the plots to fulfill the subtask requirements.



```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generar datos de muestra para múltiples series
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Datos para gráfico de dispersión (categorías)
x_scatter_adv = np.random.rand(50) * 10
y_scatter_adv = np.random.rand(50) * 10
colors_scatter = np.random.rand(50)
sizes_scatter = np.random.rand(50) * 80 + 20 # Tamaños entre 20 y 100

# 2. Crear una figura con al menos dos subgráficos (1 fila, 2 columnas)
fig_adv, axes_adv = plt.subplots(1, 2, figsize=(16, 6))

# 3. Primer subgráfico: Gráfico de líneas con múltiples series personalizadas
axes_adv[0].plot(x, y1, color='blue', linestyle='-', marker='o', markersize=4, label='Seno (sin(x))')
axes_adv[0].plot(x, y2, color='red', linestyle='--', marker='x', markersize=4, label='Coseno (cos(x))')
axes_adv[0].plot(x, y3, color='green', linestyle=':', marker='^', markersize=4, label='Producto (sin(x)*cos(x))')

axes_adv[0].set_xlabel('Ángulo (radianes)')
axes_adv[0].set_ylabel('Valor de la Función')
axes_adv[0].set_title('Gráfico de Líneas con Múltiples Series')
axes_adv[0].legend()
axes_adv[0].grid(True)

# 4. Segundo subgráfico: Gráfico de dispersión con personalización avanzada
scatter = axes_adv[1].scatter(x_scatter_adv, y_scatter_adv, c=colors_scatter, s=sizes_scatter, cmap='viridis', alpha=0.7)
axes_adv[1].set_xlabel('Variable X')
axes_adv[1].set_ylabel('Variable Y')
axes_adv[1].set_title('Gráfico de Dispersión Avanzado (Tamaño y Color)')
cbar = fig_adv.colorbar(scatter, ax=axes_adv[1])
cbar.set_label('Intensidad de Color')
axes_adv[1].grid(True)

# 5. Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous step prepared the necessary data. Now, I will create a figure with three subplots and then generate a scatter plot with a regression line using `sns.regplot()`, a distribution plot using `sns.histplot()`, and a count plot using `sns.countplot()`, all with appropriate labels and titles, and finally display them.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 2. Generate sample data
# Data for scatter plot with regression line
np.random.seed(42) # for reproducibility
x_reg = np.random.rand(100) * 10
y_reg = 2 * x_reg + np.random.randn(100) * 5
df_reg = pd.DataFrame({'X': x_reg, 'Y': y_reg})

# Data for distribution plot
data_dist = np.random.normal(loc=0, scale=1, size=1000)

# Data for count plot (categorical)
categories = ['A', 'B', 'C', 'D']
data_cat = np.random.choice(categories, size=500, p=[0.2, 0.3, 0.4, 0.1])
df_cat = pd.DataFrame({'Category': data_cat})

# 3. Create a figure with at least three subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 4. First subplot: Scatter plot with regression line
sns.regplot(x='X', y='Y', data=df_reg, ax=axes[0], scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
axes[0].set_xlabel('Variable X')
axes[0].set_ylabel('Variable Y')
axes[0].set_title('Diagrama de Dispersión con Regresión')

# 5. Second subplot: Distribution plot
sns.histplot(data_dist, kde=True, ax=axes[1], color='skyblue')
axes[1].set_xlabel('Valor')
axes[1].set_ylabel('Frecuencia / Densidad')
axes[1].set_title('Gráfico de Distribución (Histograma y KDE)')

# 6. Third subplot: Count plot
sns.countplot(x='Category', data=df_cat, ax=axes[2], palette='viridis')
axes[2].set_xlabel('Categoría')
axes[2].set_ylabel('Conteo')
axes[2].set_title('Gráfico de Conteo de Categorías')

# 7. Adjust layout and show plots
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because the `palette` argument in `sns.countplot` was used without assigning the `x` variable to `hue`. To resolve this and prevent future warnings, I will explicitly set `hue='Category'` and `legend=False` in the `sns.countplot` function.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 2. Generate sample data
# Data for scatter plot with regression line
np.random.seed(42) # for reproducibility
x_reg = np.random.rand(100) * 10
y_reg = 2 * x_reg + np.random.randn(100) * 5
df_reg = pd.DataFrame({'X': x_reg, 'Y': y_reg})

# Data for distribution plot
data_dist = np.random.normal(loc=0, scale=1, size=1000)

# Data for count plot (categorical)
categories = ['A', 'B', 'C', 'D']
data_cat = np.random.choice(categories, size=500, p=[0.2, 0.3, 0.4, 0.1])
df_cat = pd.DataFrame({'Category': data_cat})

# 3. Create a figure with at least three subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 4. First subplot: Scatter plot with regression line
sns.regplot(x='X', y='Y', data=df_reg, ax=axes[0], scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
axes[0].set_xlabel('Variable X')
axes[0].set_ylabel('Variable Y')
axes[0].set_title('Diagrama de Dispersión con Regresión')

# 5. Second subplot: Distribution plot
sns.histplot(data_dist, kde=True, ax=axes[1], color='skyblue')
axes[1].set_xlabel('Valor')
axes[1].set_ylabel('Frecuencia / Densidad')
axes[1].set_title('Gráfico de Distribución (Histograma y KDE)')

# 6. Third subplot: Count plot
sns.countplot(x='Category', data=df_cat, ax=axes[2], hue='Category', palette='viridis', legend=False)
axes[2].set_xlabel('Categoría')
axes[2].set_ylabel('Conteo')
axes[2].set_title('Gráfico de Conteo de Categorías')

# 7. Adjust layout and show plots
plt.tight_layout()
plt.show()
```

## Visualizaciones estadísticas con Seaborn (avanzadas)

### Subtask:
Crea visualizaciones estadísticas más sofisticadas con Seaborn, como mapas de calor (heatmaps), pair plots y gráficos de violín, para explorar relaciones complejas en los datos.


**Reasoning**:
First, I need to import the necessary libraries (`seaborn`, `matplotlib.pyplot`, `numpy`, `pandas`) and then generate sample data suitable for advanced Seaborn visualizations, including a Pandas DataFrame with numerical columns for heatmaps and pair plots, and a categorical column for violin plots.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 2. Generate sample data for advanced visualizations
np.random.seed(42) # for reproducibility

# Data for heatmap and pairplot (numerical columns)
data_advanced = {
    'Feature_A': np.random.rand(100) * 10,
    'Feature_B': np.random.rand(100) * 10 + np.random.rand(100) * 2,
    'Feature_C': np.random.rand(100) * 5 - np.random.rand(100) * 3,
    'Feature_D': np.random.rand(100) * 10 + 5,
    'Feature_E': np.random.rand(100) * 7
}
df_advanced = pd.DataFrame(data_advanced)

# Add some correlation for better heatmap visualization
df_advanced['Feature_B'] = df_advanced['Feature_A'] * 0.7 + np.random.randn(100) * 1.5
df_advanced['Feature_C'] = df_advanced['Feature_A'] * -0.5 + np.random.randn(100) * 1

# Add a categorical column for violinplot and hue in pairplot
categories = ['Group 1', 'Group 2', 'Group 3']
df_advanced['Category'] = np.random.choice(categories, size=100, p=[0.4, 0.3, 0.3])

print("Sample DataFrame generated for advanced Seaborn visualizations:")
print(df_advanced.head())
```

**Reasoning**:
The previous step successfully generated the `df_advanced` DataFrame. Now, I will create a figure with two subplots for the heatmap and violin plot, calculate the correlation matrix for the heatmap, generate both plots with appropriate customizations (annotations, colormaps, titles, labels), and then create a separate `pairplot` with a general title to fulfill the subtask requirements.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots for heatmap and violinplot
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# 3. Create a heatmap of correlation
# Select only numerical columns for correlation calculation
numerical_cols = df_advanced.select_dtypes(include=np.number).columns
corr_matrix = df_advanced[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=axes[0])
axes[0].set_title('Mapa de Calor de Correlación de Características')

# 5. Create a violinplot
sns.violinplot(x='Category', y='Feature_A', data=df_advanced, ax=axes[1], palette='pastel')
axes[1].set_xlabel('Categoría')
axes[1].set_ylabel('Valor de Feature_A')
axes[1].set_title('Gráfico de Violín de Feature_A por Categoría')

# 6. Adjust layout for heatmap and violinplot
plt.tight_layout()
plt.show()

# 4. Generate a pairplot (creates its own figure)
pair_plot = sns.pairplot(df_advanced, hue='Category', palette='viridis')
pair_plot.fig.suptitle('Pair Plot de Características por Categoría', y=1.02) # y adjusts title position
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because the `palette` argument in `sns.violinplot` was used without assigning the `x` variable to `hue`. To resolve this and prevent future warnings, I will explicitly set `hue='Category'` and `legend=False` in the `sns.violinplot` function.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots for heatmap and violinplot
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# 3. Create a heatmap of correlation
# Select only numerical columns for correlation calculation
numerical_cols = df_advanced.select_dtypes(include=np.number).columns
corr_matrix = df_advanced[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=axes[0])
axes[0].set_title('Mapa de Calor de Correlación de Características')

# 5. Create a violinplot
sns.violinplot(x='Category', y='Feature_A', data=df_advanced, ax=axes[1], hue='Category', palette='pastel', legend=False)
axes[1].set_xlabel('Categoría')
axes[1].set_ylabel('Valor de Feature_A')
axes[1].set_title('Gráfico de Violín de Feature_A por Categoría')

# 6. Adjust layout for heatmap and violinplot
plt.tight_layout()
plt.show()

# 4. Generate a pairplot (creates its own figure)
pair_plot = sns.pairplot(df_advanced, hue='Category', palette='viridis')
pair_plot.fig.suptitle('Pair Plot de Características por Categoría', y=1.02) # y adjusts title position
plt.show()
```

## Gráficos interactivos con Plotly

### Subtask:
Desarrolla y muestra un gráfico interactivo utilizando Plotly, como un diagrama de dispersión o un gráfico de líneas, para permitir la exploración dinámica de los datos.


**Reasoning**:
First, I will import the necessary libraries, `plotly.graph_objects` for interactive plotting and `numpy` for generating sample data. Then, I will generate random data for x and y coordinates. Finally, I will create and display an interactive scatter plot with a title to allow dynamic exploration of the data, as requested.



```python
import plotly.graph_objects as go
import numpy as np

# 2. Generate sample data for an interactive scatter plot
np.random.seed(42) # for reproducibility
x_interactive = np.random.rand(100) * 10
y_interactive = np.random.rand(100) * 10 + np.sin(x_interactive) * 2

# 3. Create an interactive scatter plot using go.Figure() and go.Scatter()
fig = go.Figure(
    data=[go.Scatter(
        x=x_interactive,
        y=y_interactive,
        mode='markers', # 4. Set the mode to 'markers' for individual points
        marker=dict(
            size=8,
            color=np.random.randn(100), # Color points based on another random value
            colorscale='Viridis',
            showscale=True
        )
    )]
)

# 5. Add a title to the plot
fig.update_layout(
    title_text='Gráfico de Dispersión Interactivo con Plotly',
    xaxis_title='Variable X',
    yaxis_title='Variable Y'
)

# 6. Display the interactive plot
fig.show()
```
